## 3. Implement Linear Regression

*Note: You can use the [dir()](https://www.geeksforgeeks.org/python-dir-function/) function to see the methods you need.*

```python
from sklearn.linear_model import LinearRegression, Ridge, RidgeCV, Lasso, LassoCV
from sklearn.preprocessing import StandardScaler, MinMaxScaler, RobustScaler
```

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score

def train_val(y_train, y_train_pred, y_test, y_pred, model_name):
    
    scores = {model_name+"_train": {"R2" : r2_score(y_train, y_train_pred),
    "mae" : mean_absolute_error(y_train, y_train_pred),
    "mse" : mean_squared_error(y_train, y_train_pred),                          
    "rmse" : np.sqrt(mean_squared_error(y_train, y_train_pred))},
    
    model_name+"_test": {"R2" : r2_score(y_test, y_pred),
    "mae" : mean_absolute_error(y_test, y_pred),
    "mse" : mean_squared_error(y_test, y_pred),
    "rmse" : np.sqrt(mean_squared_error(y_test, y_pred))}}
    
    return pd.DataFrame(scores)
```

NOT: Farklı scaling methodlar denenebilir.

```python
model_linear = LinearRegression()
```

```python
# Without scaling

model_linear.fit(X_train, y_train)
```

```python
model_linear.intercept_
```

```python
y_test_pred= model_linear.predict(X_test)
y_test_pred
```

```python
y_train_pred = model_linear.predict(X_train)
y_train_pred
```

```python
eval_without_scaling = train_val(y_train, y_train_pred, y_test, y_test_pred, "Linear")
eval_without_scaling
```

```python
# With StandartScaler

scaler = StandardScaler()
```

```python
y_train_pred = model_linear.predict(Xz_train)
y_train_pred
```

```python
train_val(y_train, y_train_pred, y_test, y_test_pred, "Linear_MinMaxScaler")
```

```python
y_train_pred = model_linear.predict(Xz_train)
y_train_pred
```

```python
# With RobustScaler

scaler = RobustScaler()
```
