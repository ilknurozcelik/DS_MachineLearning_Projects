# Raisin Class Prediction

Data Set Information:

Images of Kecimen and Besni raisin varieties grown in Turkey were obtained with CVS. A total of 900 raisin grains were used, including 450 pieces from both varieties. These images were subjected to various stages of pre-processing and 7 morphological features were extracted. These features have been classified using three different artificial intelligence techniques.


Attribute Information:

1. Area: Gives the number of pixels within the boundaries of the raisin.
2. Perimeter: It measures the environment by calculating the distance between the boundaries of the raisin and the pixels around it.
3. MajorAxisLength: Gives the length of the main axis, which is the longest line that can be drawn on the raisin.
4. MinorAxisLength: Gives the length of the small axis, which is the shortest line that can be drawn on the raisin.
5. Eccentricity: It gives a measure of the eccentricity of the ellipse, which has the same moments as raisins.
6. ConvexArea: Gives the number of pixels of the smallest convex shell of the region formed by the raisin.
7. Extent: Gives the ratio of the region formed by the raisin to the total pixels in the bounding box.
8. Class: Kecimen and Besni raisin.

https://archive.ics.uci.edu/ml/datasets/Raisin+Dataset

## Model Performance
*Check the accuracy of the model using the train and test set by given eval_metric function.

```python
def eval_metric(model, X_train, y_train, X_test, y_test):

    """ to get the metrics for the model """

    y_train_pred = model.predict(X_train)
    y_pred = model.predict(X_test)
    
    print("Test_Set")
    print(confusion_matrix(y_test, y_pred))
    print(classification_report(y_test, y_pred))
    print()
    print("Train_Set")
    print(confusion_matrix(y_train, y_train_pred))
    print(classification_report(y_train, y_train_pred))
```

```python
eval_metric(log_model_pipe, X_train, y_train, X_test, y_test)
```

### GridSearchCV
* Use the GridSearchCV function to find the best parameters for the model.

```python
from sklearn.model_selection import GridSearchCV
log_model=LogisticRegression()
# pipeline for logistic regression
log_model_pipe = Pipeline([("scaler",scaler),("log_model",log_model)]) 
# l1: Lasso, l2: Ridge
penalty = ["l1", "l2"] 
# to get 20 values of C between -1 and 5
C = np.logspace(-1, 5, 20) 
# balanced: class weights are balanced, None: no class weights
class_weight= ["balanced", None] 
# to get 4 values of solver
solver = ["lbfgs", "liblinear", "sag", "saga"] 
# to get all the combinations of penalty, C, class_weight and solver
param_grid = {"log_model__penalty" : penalty,
              "log_model__C" : [C,1],
              "log_model__class_weight":class_weight,
              "log_model__solver":solver} 

# to get the best model
grid_model = GridSearchCV(estimator=log_model_pipe,
                          param_grid=param_grid,
                          cv=10,
                          scoring = 'accuracy',       
                          n_jobs = -1) 
```

### Elbow Method for Choosing Reasonable K Values
* Use the elbow method to find the optimal number of neighbors.

```python
test_error_rates = []


for k in range(1,30):
    knn_model = KNeighborsClassifier(n_neighbors=k)
    operations =  [('scaler',StandardScaler()),('knn',knn_model)]
    knn_pipe_model = Pipeline(operations)
    
    knn_pipe_model.fit(X_train,y_train) 
   
    y_pred_test = knn_pipe_model.predict(X_test)
    
    test_error = 1 - accuracy_score(y_test,y_pred_test)  # test setindeki hatay bulmak iin
    
    test_error_rates.append(test_error)
    
```

```python
plt.figure(figsize=(15,8))
plt.plot(range(1,30), test_error_rates, color='blue', linestyle='--', marker='o',
         markerfacecolor='red', markersize=10)

plt.title('Error Rate vs. K Value')
plt.xlabel('K_values')
plt.ylabel('Error Rate')
plt.hlines(y=0.139, xmin = 0, xmax = 30, colors= 'r', linestyles="--")
plt.hlines(y=0.144, xmin = 0, xmax = 30, colors= 'r', linestyles="--");
```

### Overfitting and underfitting control

```python
test_error_rates = []
train_error_rates = []


for k in range(1,30):
    knn_model = KNeighborsClassifier(n_neighbors=k)
    operations =  [('scaler',StandardScaler()),('knn',knn_model)]
    knn_pipe_model = Pipeline(operations)
    
    knn_pipe_model.fit(X_train,y_train) 
   
    y_pred_test = knn_pipe_model.predict(X_test)
    y_pred_train = knn_pipe_model.predict(X_train)
    
    test_error = 1 - accuracy_score(y_test,y_pred_test)
    train_error = 1 - accuracy_score(y_train,y_pred_train)
    test_error_rates.append(test_error)
    train_error_rates.append(train_error)
```

```python
plt.figure(figsize=(15,8))
plt.plot(range(1,30), test_error_rates, color='blue', linestyle='--', marker='o',
         markerfacecolor='red', markersize=10)
plt.plot(range(1,30), train_error_rates, color='green', linestyle='--', marker='o',
         markerfacecolor='blue', markersize=10)
plt.title('Error Rate vs. K Value')
plt.xlabel('K_values')
plt.ylabel('Error Rate')
plt.hlines(y=0.139, xmin = 0, xmax = 30, colors= 'r', linestyles="--")
plt.hlines(y=0.144, xmin = 0, xmax = 30, colors= 'r', linestyles="--")
plt.legend(["test", "train"]);
```

```python
## Allen Hoca'dan

def elbow_method(rang=range(1,30)):
    test_error_rates = []
    train_error_rates = []

    for k in rang: 
        # {'n_neighbors': 21, 'p': 2, 'weights': 'uniform'}
        knn_model = KNeighborsClassifier(n_neighbors=k)
        scaler = StandardScaler()  
        knn_pipe_model = Pipeline([
            ("scaler",scaler),
            ("knn_model",knn_model)
        ]) 
        knn_pipe_model.fit(X_train,y_train) 
        y_pred_test = knn_pipe_model.predict(X_test)
        y_pred_train = knn_pipe_model.predict(X_train)
        
        # test setindeki hatay bulmak iin
        test_error = 1 - accuracy_score(y_test, y_pred_test)  
        train_error = 1 - accuracy_score(y_train, y_pred_train)
        test_error_rates.append(test_error)
        train_error_rates.append(train_error)
        
    return test_error_rates, train_error_rates
```

```python
test_error_rates, train_error_rates = elbow_method()
```

```python
plt.figure(figsize=(15,8))
plt.plot(
    range(1,30), test_error_rates, 
    color='blue', linestyle='--', lw=0.3, 
    marker='o',markerfacecolor='red', markersize=10,
)      
for x,y in enumerate(test_error_rates):
    plt.annotate(x+1, (x+0.5, y), size=10, weight='heavy') 
    
plt.plot(
    range(1,30), train_error_rates, 
    color='green', linestyle='--', lw=0.3,
    marker='o', markerfacecolor='blue', markersize=10
)     
for x,y in enumerate(train_error_rates):
    plt.annotate(x+1, (x+0.5, y), size=10, weight='heavy')
    
# first 10 point, next 10
plt.hlines(y=min(test_error_rates[:10]), xmin=0, xmax=30, colors= 'r', linestyles="--", label='min first 10')
plt.hlines(y=min(test_error_rates[10:20]), xmin=0, xmax=30, colors= 'r', linestyles=":", label='min next 10')
plt.hlines(y=max(train_error_rates[:10]), xmin=0, xmax=30, colors= 'b', linestyles="--", label='min first 10')
plt.hlines(y=max(train_error_rates[10:20]), xmin=0, xmax=30, colors= 'b', linestyles=":", label='min next 10')
plt.legend(loc='lower right')
plt.title('Error Rate vs. K Value')
plt.xlabel('K_values')
plt.ylabel('Error Rate')
plt.show()
```


### Gridsearch Method for Choosing Reasonable K Values
* Use the GridSearchCV function to find the best parameters for the model. this method is brute force method to find the best parameters, and it takes a long time to run for KNN. You will get different k values because of you will use different parameters, and also gridsearch method tries to find minimum error. Sometimes it get larger k values for optimal but for  sake of reduce complexity we use the elbow method to find the optimal k value.

```python
from sklearn.model_selection import GridSearchCV

operations = [("scaler", StandardScaler()), ("knn", KNeighborsClassifier())]
knn_pipe_model = Pipeline(operations)

param_grid = {'knn__n_neighbors': range(1,30), 'knn__p': [1,2], "knn__weights" : ["uniform", "distance"] }

# to get the best model
knn_pipe_grid_model = GridSearchCV(estimator=knn_pipe_model,
                          param_grid=param_grid,
                          cv=10,
                          scoring = 'accuracy',       
                          n_jobs = -1)
```

```python
knn_pipe_grid_model.fit(X_train, y_train)
```

```python
knn_pipe_grid_model.best_params_
```
